```python
import os
import pandas as pd
import requests
```

```python
csv_file = "clean_MovieGenre.csv"
save_dir = "downloaded_images"
```

```python
def download_image(url, save_dir, imdb_id, failed_downloads):
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        # Check if the request was successful
        if response.status_code == 200:
            # Extract the filename from the URL
            filename = os.path.join(save_dir, f"{imdb_id}.jpg")
            # Save the image locally
            with open(filename, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded: {imdb_id}")
        else:
            print(f"Failed to download: {imdb_id} (Status Code: {response.status_code})")
            failed_downloads.append(imdb_id)  # Add to list of failed downloads
    except Exception as e:
        print(f"Error downloading {imdb_id}: {e}")
        failed_downloads.append(imdb_id)  # Add to list of failed downloads
```

```python
df = pd.read_csv(csv_file)
image_urls = df['Poster'].tolist() 
imdb_ids = df['imdbId'].tolist()
```

```python
failed_downloads = []

# Download images
for url, imdb_id in zip(image_urls, imdb_ids):
    download_image(url, save_dir, imdb_id, failed_downloads)
```
